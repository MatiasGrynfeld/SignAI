```python
if handPointsKeyFrames:
    normalizer = Point2Vec(4)
    hand_landmarks = normalizer.land2vec(handPointsKeyFrames)
    print(hand_landmarks)
```

```python
def create_dataframe(handLandmarks):
    flattened_data = []
    for frame in handLandmarks:
        row = []
        for hand_landmarks in frame:
            for point in hand_landmarks:
                row.append(point)
        flattened_data.append(row)
    columns = []
    for hand in range(1, 3):
        for point in range(1, 22):
            columns.append(f'h{hand}p{point}')
    df = pd.DataFrame(flattened_data, columns=columns)
    return df
```

```python
handDataFrame = create_dataframe(hand_landmarks)
handDataFrame
```
