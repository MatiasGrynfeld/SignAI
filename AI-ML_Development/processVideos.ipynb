```python
import os
from KeyFrameExtractorClass import KeyFrameExtractor
from Points2VecClass import Point2Vec
import cv2
from VideoFormaterClass import VideoFormater
```

```python
videos_path = "C:\\Users\\48519558\\Desktop\\SignAI-ML\\AI-ML_Development\\recursos\\videos"

video_files = []

for filename in os.listdir(videos_path):
    if filename.endswith('.mp4'):
        video_files.append(os.path.join(videos_path, filename))
```

```python
kfe = KeyFrameExtractor()
videoFormater = VideoFormater()
normalizer = Point2Vec(4)
dataFrames = []
for index, path in enumerate(video_files):
    video = cv2.VideoCapture(path)
    video_points = kfe.extractKeyFrames(return_frame=False, draw=False, video=video)
    landmarks = normalizer.land2vec(video_points)
    dataFrames.append(videoFormater.formatVideo(landmarks, "A", index))
```

```python
VideoFormater.concatAndExportVideos(VideoFormater, dataFrames)
```
